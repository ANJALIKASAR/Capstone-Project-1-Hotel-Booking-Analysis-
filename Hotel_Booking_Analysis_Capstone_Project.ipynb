# 2.Which is the busiest year and month ?

```python
arrival_date_year=hotel_data['arrival_date_year'].unique() # checking unique year using unique() function.
arrival_date_year
```

```python
plt.rcParams['figure.figsize']=(8,5)
year_counts = hotel_data.arrival_date_year.value_counts() ##count values of arrival date year.
year_counts.sort_index(inplace=True)
sns.barplot(x = year_counts.index, y = year_counts)
plt.title("Rate of customer ",fontsize='20')
plt.xlabel('arrival_date_year',fontsize='15')
plt.ylabel('counts',fontsize='15')
plt.show()
```

#*Analysis:* 
From the above bar plots, we can see 2016 was the busiest year compare to 2015 and 2017.

```python
##Count the visitors in each month.
plt.rcParams['figure.figsize']=(12,6)
month_sorted=['January','February','March','April','May','June','July','August','September','October','November','December']
sns.countplot(data=hotel_data ,x='arrival_date_month', palette='pastel',
              order=pd.value_counts(hotel_data['arrival_date_month']).index)##count the number of bookings made in a month using value_count function

plt.title('Busiest Month with Bookings')
plt.xlabel('Month')
plt.ylabel('Reservation Count')
plt.show()
```

#*Analysis:* 
From the above bar charts, we can see that July and August has the most number of bookings and on the other hand January has the least booked month.

#3.Which country has made more number of bookings in hotels comparing with other countries?

It seems more number of countries are in a dataset.Let's see how many countries have booked the hotel.

Ok, let's visualise this using graph.

```python
##use the countplot to visualise the data 
plt.rcParams['figure.figsize']=(15,5)
country_counts = hotel_data['country'].value_counts().sort_values(ascending=False)[0:20]##sort it in descending order maximum to minimum.
sns.countplot(x='country', hue='hotel', data=hotel_data[hotel_data['country'].isin(country_counts.index)])
plt.title("Booked Hotel type by country",fontsize='18')
plt.xlabel("Country",fontsize='15')
plt.ylabel("No. of Bookings",fontsize='15')
plt.show()
```

#6. Which mode of  booking are higher comparing with direct,online,offline and other agents?

```python
#listing different market segment in hotel booking
hotel_data['market_segment']
```

```python
#ploting a pie chart comparing with different market segments
plt.rcParams['figure.figsize']=(8,8)
group_by_dc = hotel_data.groupby('market_segment')
d1 = pd.DataFrame(round((group_by_dc.size()/hotel_data.shape[0])*100,2)).reset_index().rename(columns = {0: 'Booking_%'})
data = d1['Booking_%']
hue= 'market_segment'
labels = d1['market_segment']
plt.pie(x=data, autopct="%.2f%%",  explode=[0.02]*8 ,labels=labels, pctdistance=0.5)
plt.title("Booking % by market_segment", fontsize=14);
```

By observing the above piechart we can conclude that most number of bookings done by online mode. Here, 70% of booking done using three methods are: Online, Offline and groups compared to direct and other agents booking.
